```python
#Q1

#the functions req
def d2ydt2(t, y, v):
    z= (2*math.cos(1.2*t) - 2*y - 0.2*v)/2
    return z

def dydt(t, y, v):
    return v

#the oscillator solved using RK4 integrator
x0_1, y0_1, z0_1 = ar.RKXY(d2ydt2, dydt, 0, 2, -1, 50, 0.5, 'yes')
```

```python
#Q2

#the variables
x0=0
xf=10
y0=40
yf=200
z0=1
dx=0.01
tol=0.001

#the functions req
def d2Tdx2(x, T, t):
    return -1*0.01*(20-T)

def dTdx(x, T, t):
    return t

#BDV using Shooting method
x, T, t = ar.ShootXY(d2Tdx2, dTdx, x0, xf, y0, yf, z0, dx, tol, 'yes')

#finding closest temp to 100 in the rod
print('\n The temp at which rod is 100 +- 0.1 C is:\n')
for i in range(0,len(x)):
    if abs(T[i]-100) <= 0.1:
        print(T[i], 'C at', x[i],'m.')
```
