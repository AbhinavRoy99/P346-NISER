```python
#opening the file with value of N
f = open('input1.txt', 'r')
content = f.readlines()

#assigning variables for the loop
x=1
y=1
a=1
b=1
s=0
p=1

#reading value of N from file
for i in content:
    i = i.strip()
    i = int(i)

#while loop for odd num sum    
while a <= i:
    s=s+x
    x=x+2
    a=a+1



#while loop for factorial
while b <= i:
    p=p*y
    y=y+1
    b=b+1



#writing output to a new file
o = open("output1.txt", 'w')
o.write("Number of terms 'N' is \n")
o.write(str(i))
o.write("\n")
o.write("The sum of first 'N' odd numbers is \n")
o.write(str(s))
o.write("\n")
o.write("The factorial of 'N' is \n")
o.write(str(p))

#For notebook file
print("Number of terms 'N' is \n")
print(i)
print("The sum of first 'N' odd numbers is \n")
print(s)
print("The factorial of 'N' is \n")
print(p)
```

```python
import decimal
import math

#opening the file with values of Nadr
f = open('input2.txt', 'r')
content = f.read().split(',')

print(content)

#assigning values for Nadr from the content array
n=float(content[0])
a=float(content[1])
d=float(content[2])
r=float(content[3])

#for AP
i=1
s_ap=0
b=0
b=a

while i <= n:
    s_ap=s_ap+b
    b=b+d
    i=i+1
    


#for GP
j=1
s_gp=0
c=0
c=a

while j <= n:
    s_gp=s_gp+c
    c=c*r
    j=j+1
    


#for HP
k=1
s_hp=0
e=0
e=(1/a)

while k <= n:
    s_hp=s_hp+e
    e1=(1/e)
    e1=e1+d
    e=(1/e1)
    k=k+1
    


#writing output to a new file
o = open("output2.txt", 'w')
o.write("Number of terms 'N' is \n")
o.write(str(n))
o.write("\n")
o.write("The first term 'a' is \n")
o.write(str(a))
o.write("\n")
o.write("The common difference 'd' is \n")
o.write(str(d))
o.write("\n")
o.write("The common ratio 'r' is \n")
o.write(str(r))
o.write("\n")
o.write("The sum of AP is \n")
o.write(str(s_ap))
o.write("\n")
o.write("The sum of GP is \n")
o.write(str(s_gp))
o.write("\n")
o.write("The sum of HP is \n")
o.write(str(s_hp))

print("Number of terms 'N' is")
print(n)

print("The first term 'a' is")
print(a)

print("The common difference 'd' is")
print(d)

print("The common ratio 'r' is")
print(r)

print("The sum of AP is")
print(s_ap)

print("The sum of GP is")
print(s_gp)

print("The sum of HP is")
print(s_hp)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

s=(1/2)

#setting up a prev and next sum
s_p=0
s_n=0

#the logic here used was the loop should go on
#until we find the the prev sum and the next sum has a difference
#less than  that of 0.0001 the accuracy required
while abs(s_n) >= 0:
    s_p=s_n
    s_n=s_n+s
    s=s*(-1/2)
    plt.scatter(s,s_p)
    if abs(s_n - s_p) < 0.00001:
        break

print("Sum of the series accurate to 4 decimal places is \n")
print(s_n)

#writing output and saving plot to a new file
o = open("output3.txt", 'w')
o.write("Sum of the series accurate to 4 decimal places is \n")
o.write(str(s_n))
plt.savefig("outputplot3")
```

```python
import math
#defined the complex function
class myComplex:
    type = 'compnum'
    def __init__(self,r_n,i_n):
        self.r_n = int(r_n)
        self.i_n = int(i_n)
    
#defined the functions for complex numbers
def compadd(a1,a2):
    if a1.type == 'compnum':
        if a2.type == 'compnum':
            x = a1.r_n + a2.r_n
            y = a1.i_n + a2.i_n
            print("The sum of the complex numers is: (",x,") +  i(",y,")")
        else:
            print("These are not complex numbers.")
    else:
        print("These are not complex numbers.")

def compprod(a1,a2):
    if a1.type == 'compnum':
        if a2.type == 'compnum':
            x = (a1.r_n * a2.r_n) - (a1.i_n * a2.i_n)
            y = (a1.r_n * a2.i_n) + (a1.i_n * a2.r_n)
            print("The product of the complex numers is: (",x,") +  i(",y,")")
        else:
            print("These are not complex numbers.")
    else:
        print("These are not complex numbers.")

def compmod(a1):
    if a1.type == 'compnum':
        x = (a1.r_n * a1.r_n)
        y = (a1.i_n * a1.i_n)
        mod = math.sqrt((x*x) + (y*y))
        mod = round(mod,3)
        print("The modulus of the complex number is: |",mod,"|")
    else:
        print("This is not a complex number.")


a = myComplex(3,-2)
b = myComplex(1,2)

compadd(a,b)
compprod(a,b)
compmod(a)
compmod(b)
```
