```python
#Q1 GAUSS JORDAN

with open('m1.txt', 'r') as f1:
    lq1 = [[float(num) for num in line.split(',')] for line in f1 ]

with open('b1.txt', 'r') as f1:
    kq1 = [[float(num) for num in line.split(',')] for line in f1 ]

#printing the equations
ar.printeq(lq1,kq1,'X','Given')

ar.printhash()

#solving the system of equations using gauss jordan code
ar.gaussjordan(lq1,kq1)

ar.printhash()
```

```python
#Q1 LU DECOMPOSITION

with open('m1.txt', 'r') as f1:
    lq1 = [[float(num) for num in line.split(',')] for line in f1 ]

with open('b1.txt', 'r') as f2:
    kq1 = [[float(num) for num in line.split(',')] for line in f2 ]
    
ar.printeq(lq1,kq1,'X','Given')

ar.printhash()

#for pivoting the matrix i have used a different approach than the one discussed in class
#i was reading online and i found that we can pivot the matrix by introducing a pivoting matrix
#so my approach here is a lil different than that discussed in classroom
pq1=ar.pivotmat(lq1)

#pivot matrix is also useful for pivoting the other matrix
kq1=ar.crossmat(pq1,kq1)

#finding the L and U using library function
L1,U1=ar.LUdecomposition(lq1,pq1)

ar.printhash()

#forward subsitutuion
ar.printeq(L1,kq1,'Y','For forward subsitution the given')

kq11=ar.forwardsub(L1,kq1)
print('\n After forward subsitution Matrix Y is: \n', kq11)

ar.printhash()
#backward subsitution
ar.printeq(U1,kq11,'X','For backward subsitution the given')

kq12=ar.backwardsub(U1,kq11)
print('\n After backward subsitution the answer matrix X is: \n',kq12)

ar.printhash()

#i find that answers for both gauss jordan and LU decomposition are coming wildly different
#i am yet to find a reason to it
#i'm confident that it is not because i've written code wrong
#since i have independently tested both codes for other system of equations as well
#i am still trying to find a reason for this
```

```python
#Q2 CHOLESKY

with open('m2.txt', 'r') as f1:
    l = [[int(num) for num in line.split(',')] for line in f1 ]

with open('b2.txt', 'r') as f2:
    k = [[int(num) for num in line.split(',')] for line in f2 ]

ar.printeq(l,k,'X','Given')

#check for symmetry of matrix
ar.checksym(l)

ar.printhash()

#cholesky decomposition using library function
l,lt = ar.CholeskyD(l)

ar.printhash()
#forward subsitution
ar.printeq(l,k,'Y','For forward subsitution the given')

k1=ar.forwardsub(l,k)
print('\n After forward subsitution Matrix Y is: \n', k1)

ar.printhash()
#backward subsitution
ar.printeq(lt,k1,'X','For backward subsitution the given')

k2=ar.backwardsub(lt,k1)
print('\n After backward subsitution the answer matrix X is: \n',k2)

ar.printhash()
```

```python
#Q2 GAUSS SEIDEL

with open('m2.txt', 'r') as f1:
    lq2 = [[int(num) for num in line.split(',')] for line in f1 ]

with open('b2.txt', 'r') as f2:
    kq2 = [[int(num) for num in line.split(',')] for line in f2 ]

ar.printeq(lq2,kq2,'X','Given')

ar.printhash()

k1q2=[]
for i in range(0,len(lq2)):
    k1q2.append(kq2[i][0])

#(: guess is taken to be 10 since 1 is the answer itself :)
x= [10]*len(lq2)

print("\n Guess: \n",x)


ar.GaussSeidel(lq2,x,k1q2,1000,0.000001)

ar.printhash()
```

```python
#Q3 JACOBI & GAUSS-SEIDEL

with open('m3.txt', 'r') as f1:
    lq3 = [[int(num) for num in line.split(',')] for line in f1 ]

with open('b3.txt', 'r') as f2:
    kq3 = [[int(num) for num in line.split(',')] for line in f2 ]

ar.printeq(lq3,kq3,'X','Given')

lq31,kq31=ar.diagonaldom(lq3,kq3)

ar.printhash()

kq311=[]
for i in range(0,len(kq3)):
    kq311.append(kq31[i][0])

ar.JacobiM(lq31,kq311,100,0.000001)

ar.printhash()

x= [1]*len(kq311)

print("\n Guess: \n",x)

ar.GaussSeidel(lq31,x,kq311,1000,0.000001)

ar.printhash()
```

```python
#Q1 LU DECOMPOSITION

with open('m3.txt', 'r') as f1:
    lq3 = [[int(num) for num in line.split(',')] for line in f1 ]

with open('b3.txt', 'r') as f2:
    kq3 = [[int(num) for num in line.split(',')] for line in f2 ]

ar.printeq(lq3,kq3,'X','Given')

ar.printhash()

pq3=ar.pivotmat(lq3)

kq3=ar.crossmat(pq3,kq3)

L3,U3=ar.LUdecomposition(lq3,pq3)

ar.printhash()

ar.printeq(L3,kq3,'Y','For forward subsitution the given')

kq31=ar.forwardsub(L3,kq3)
print('\n After forward subsitution Matrix Y is: \n', kq31)

ar.printhash()

ar.printeq(U3,kq31,'X','For backward subsitution the given')

kq32=ar.backwardsub(U3,kq31)
print('\n After backward subsitution the answer matrix X is: \n',kq32)

ar.printhash()
```
