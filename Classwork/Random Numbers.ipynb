```python
%matplotlib inline
import matplotlib.pyplot as plt

n = 0.1
c = 3.5

for i in range(1,500):
    n = c*n*(1-n)
    print(n)
    x=i
    y=n
    plt.scatter(x, y)
```

```python
a = 1103515245
c = 12345
m = 32768
n = 10
p = 12
z = 0
t = 0
for i in range(1,100000000):
    n = ((a*n+c)%m)/m
    p = ((a*p+c)%m)/m
    x=p*p
    y=n*n
    t=t+1
    if (x+y)<1:
        z=z+1  

area = 4*(z/t)
print(area)   
```
