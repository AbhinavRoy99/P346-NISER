```python
import toolsar
import math
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib notebook

x=0
y=0
z=0

t=0
a=0

arrx=[]
arry=[]
arrz=[]

#seeds
r1=100
r2=81
r3=64

for i in range(0,10000):
    #changed parameters for more accurate estimate of sphere's vol
    x = toolsar.rnumlt1(1103515245,12345,32768,r1)
    r1 = x
    
    y = toolsar.rnumlt1(1103515245,52143,32768,r2)
    r2 = y
    
    z = toolsar.rnumlt1(1103515245,32514,32768,r3)
    r3 = z
    
    t=t+1
    
    rad = math.sqrt((x*x)+(y*y)+(z*z))
    
    if rad<1:
        a=a+1
        arrx.append(x)
        arry.append(y)
        arrz.append(z)
        

vol = (a/t)

sphere = plt.axes(projection='3d')
sphere.scatter3D(arrx, arry, arrz, c=arrz)


print('The volume estimate by throwing method is:',vol)
```

```python
import toolsar
import math
import matplotlib.pyplot as plt
%matplotlib notebook

x = 0
y = 0

x1 = 0
y1 = 0

arr_x1=[]
arr_y1=[]

r = 0
sum_r = 0

#seeds
r1 = 10
r2 = 10


for i in range(0,300):
    t1 = toolsar.rnumlt1(1103515245,12345,32768,r1)
    r1=t1
    t1 = 2*(math.pi)*t1
    
    t2 = toolsar.rnumlt1(1103515245,54321,32768,r2)
    r2=t2
    t2 = 2*(math.pi)*t2
    
    x1 = x1 + 1*math.cos(t1)
    y1 = y1 + 1*math.sin(t2)
    
    arr_x1.append(x1)
    arr_y1.append(y1)
    
    r = (math.cos(t1)**2)+(math.sin(t2)**2)
    sum_r = sum_r + r

plt.plot(arr_x1, arr_y1, 'o-')
plt.savefig('randomwalk300.png')

netd1 = math.sqrt(((arr_x1[299]**2)-(x**2))+((arr_y1[299]**2)-(y**2)))
print('The net displacement of the random walk is:',netd1)

rms_1 = math.sqrt(sum_r/300)
print('The RMS distance covered by the particle is:',rms_1)
```
