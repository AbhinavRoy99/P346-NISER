```python
import toolsar
import math
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
%matplotlib notebook

#changed parameters for more accurate estimate of sphere's vol

result1 = toolsar.rnumlt1([10000,1103515245,12345,32768,100])

result2 = toolsar.rnumlt1([10000,1103515245,52143,32768,81])

result3 = toolsar.rnumlt1([10000,1103515245,32514,32768,64])

t=0
a=0

arrx=[]
arry=[]
arrz=[]


for i in range(0,10000):
    x = result1[i]
    y = result2[i]
    z = result3[i]
    
    t=t+1
    
    rad = math.sqrt((x*x)+(y*y)+(z*z))
    
    if rad<1:
        a=a+1
        arrx.append(x)
        arry.append(y)
        arrz.append(z)
        

vol = (a/t)

sphere = plt.axes(projection='3d')
sphere.scatter3D(arrx, arry, arrz, c=arrz)


print('The volume estimate by throwing method is:',vol)
```

```python
import toolsar
import math
import matplotlib.pyplot as plt
%matplotlib notebook

result1 = toolsar.rnumlt1([300,1103515245,12345,32768,10])

x = 0
y = 0

x1 = 0
y1 = 0

arr_x1=[]
arr_y1=[]

r = 0
sum_r = 0


for i in range(0,300):
    t1 = result1[i]
    t1 = 2*(math.pi)*t1
    
    x1 = x1 + 1*math.cos(t1)
    y1 = y1 + 1*math.sin(t1)
    
    arr_x1.append(x1)
    arr_y1.append(y1)
    
    r = (((x1**2)-(x**2))+((y1**2)-(y**2)))
    sum_r = sum_r + r

plt.plot(arr_x1, arr_y1, 'o-')
plt.savefig('randomwalk300.png')

netd1 = math.sqrt(((arr_x1[299]**2)-(x**2))+((arr_y1[299]**2)-(y**2)))
print('The net displacement of the random walk is:',netd1)

rms_1 = math.sqrt(sum_r/300)
print('The RMS distance covered by the particle is:',rms_1)
```
